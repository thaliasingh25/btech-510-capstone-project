# 1. Locating the TP53 Variant using the Variant ID rs11540652 from Ensembl

```python
!echo "Getting the location of tp53 variant id rs11540652" 
!curl -s 'https://rest.ensembl.org/variation/human' -H 'Content-type:application/json' \
-H 'Accept:application/json' -X POST -d '{ "ids" : ["rs11540652"] }' | jq -r '"Variant rs11540652 location: \(.rs11540652.mappings[0].location)"'
```
